```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import missingno as msno

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

from catboost import CatBoostClassifier

import shap
shap.initjs()

import dill

import warnings
warnings.simplefilter("ignore")
```

```python
train = pd.read_csv('data/aug_train.csv')
```

```python
train.head(2)
```

```python
features = ['enrollee_id', 'city', 'city_development_index', 'training_hours']
train_t = train.drop(features, axis=1)
train_t.head(2)
```

```python
msno.matrix(train_t)
```

```python
train_t.target.value_counts()
```

```python
data = train_t
cat_feature = list(data.select_dtypes(include=np.object).columns)
num_feature = 'target'

for cat in cat_feature:
    plt.figure(figsize = (12, 3))
    sns.pointplot(x=cat, y=num_feature, data=data, capsize=.1)
    plt.show()
    print(data[cat].value_counts())
    print(f'NaN = {data[cat].isnull().sum()}')
    print('*'*15)
```

```python
# минимальная обработка данных
train_t.loc[(train_t['gender'] == 'Other') | (train_t['gender'].isnull()), 'gender'] = 'Female'

train_t.loc[train_t['enrolled_university'].isnull(), 'enrolled_university'] = 'no_enrollment'

train_t.loc[train_t['education_level'].isnull(), 'education_level'] = 'High School'

train_t.loc[train_t['major_discipline'].isnull(), 'major_discipline'] = 'No Major'

train_t['experience'] = train_t['experience'].map({np.nan:'<2', '<1':'<2', '1':'<2', \
        '2':'2-5', '3':'2-5', '4':'2-5', '5':'5-8', '6':'5-8', '7':'5-8', \
        '8':'8-15', '9':'8-15', '10':'8-15', '11':'8-15', '12':'8-15', '13':'8-15', '14':'8-15', '15':'8-15', \
        '16':'>15', '17':'>15', '18':'>15', '19':'>15', '20':'>15', '>20': '>15'})

train_t['company_size'] = train_t['company_size'].map({np.nan:'<100', '<10':'<100', '10/49':'<100', \
        '50-99':'<100', '100-500':'100-1000', '500-999':'100-1000', \
        '1000-4999':'1000-10000', '5000-9999':'1000-10000', '10000+':'>10000'})

train_t.loc[train_t['company_type'].isnull(), 'company_type'] = 'Other'

train_t['last_new_job'] = train_t['last_new_job'].map({np.nan:'never', 'never':'never', '1':'1-4', '2':'1-4', \
                                                       '3':'1-4', '4':'1-4', '>4':'>4'})
```

```python
X = train_t.drop('target', axis=1)
y = train_t['target']

X_train, X_test = train_test_split(X, train_size=0.6, shuffle=True, random_state=42)
y_train, y_test = train_test_split(y, train_size=0.6, shuffle=True, random_state=42)

#save test
X_test.to_csv("data/X_test.csv", index=None)
y_test.to_csv("data/y_test.csv", index=None)

X_train.shape, X_test.shape
```

```python
%%time 

cat_features = list(X_train.select_dtypes(include=np.object).columns)

params_cb = { "n_estimators": 700,
                "loss_function": "Logloss",
                "eval_metric": "AUC",
                "task_type": "CPU",
                "max_bin": 20,
                "verbose": 50,
                "max_depth": 6,
                "l2_leaf_reg": 100,
                "early_stopping_rounds": 50,
                "thread_count": 6,
                "cat_features": cat_features,
                "random_seed": 42}

model_cb = CatBoostClassifier(**params_cb)

model_cb.fit(X_train, y_train, 
                 eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
pred = model_cb.predict_proba(X_test)[:,1]
roc_auc_score(y_test, pred)
```

```python
model = model_cb

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_train)

shap.summary_plot(shap_values, X_train, max_display=73)
```

```python
with open("models/CatBoost_model.dill", "wb") as f:
    dill.dump(model_cb, f)
```
